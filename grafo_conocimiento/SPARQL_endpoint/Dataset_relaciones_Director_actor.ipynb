Se crea dataset con relaciones de Actores y peliculas

```python
actors = df.drop(['movie','uri','director'],axis=1)
```

```python
actors.head(10)
```

```python
len(actors)
```

```python
# dropping NaN values 
actors=actors.dropna()
```

```python
len(actors)
```

```python
actors.reset_index(drop=True).head(10)
```

```python
#Se crea Dataframe para guardar valores de actores
d = {'id': [],'starring': []}
df_r = pd.DataFrame(data=d)
for x in range(len(actors)):
    #Se crea lista de los valores separados por coma
    lista = actors.starring.iloc[x].split(',')
    #Se convierte lista en array
    array = np.asarray(lista)
    id_m = np.full_like(array, actors.id.iloc[x])
    n_array = np.array([id_m, array])
    n_array = n_array.transpose()
    #n_array = np.append(array, id_m,axis=1)
    #Se adiciona al dataset el array
    df_r = df_r.append(pd.DataFrame(n_array, columns=['id','starring']), ignore_index=True)
```

```python
df_r
```

```python
csv_name = "r_actors.csv"
df_r.to_csv(csv_name, index=False, sep=';',encoding='utf-8')
```
